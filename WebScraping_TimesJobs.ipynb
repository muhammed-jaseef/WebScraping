```python
from bs4 import BeautifulSoup
import requests
```

```python
weblink = "https://www.timesjobs.com/candidate/job-search.html?from=submit&luceneResultSize=25&txtKeywords=data%20science&cboWorkExp1=0&postWeek=60&searchType=personalizedSearch&actualTxtKeywords=Data%20Science&searchBy=0&rdoOperator=OR&txtLocation=Bengaluru/%20Bangalore&pDate=I&sequence=1&startPage=1"
html_text = requests.get(weblink).text
html_text
```

```python
soup = BeautifulSoup(html_text, 'lxml')
soup
```

```python
jobs = soup.find_all('li', class_="clearfix job-bx wht-shd-bx")
print(jobs)
```

```python
# to get company name
#company_names=[]
for job in jobs:
  company_name = job.find('h3', class_="joblist-comp-name").text
  #company_names.append(company_name.text)
  print(company_name)
```

```python
# to remove spaces and new lines
for job in jobs:
  company_name = job.find('h3',class_="joblist-comp-name").text.replace("\t\n","").replace("(More Jobs)", "").strip()
  print(company_name)
```

```python
# to get key skills
for job in jobs:
  skills_req = job.find('span',class_="srp-skills").text.replace("\t\n","").strip()
  print(skills_req)
```

```python
# to get date posted
for job in jobs:
  date_posted = job.find('span', class_ = "sim-posted").text.replace("\t\n","").strip()
  print(date_posted)
```

```python
# to get link to apply
for job in jobs:
  links_to_apply = job.header.h2.a['href']
  print(links_to_apply)
```

```python
import pandas as pd
empty_df = pd.DataFrame(columns=['Company', 'Skills', 'Date posted', 'Apply link'])
empty_df
```

```python
empty_df = pd.DataFrame(columns=['Company', 'Skills', 'Date posted', 'Apply link'])

for job in jobs:
  company_name = job.find('h3',class_="joblist-comp-name").text.replace("\t\n","").replace("(More Jobs)", "").strip()
  skills_req = job.find('span',class_="srp-skills").text.replace("\t\n","").strip()
  date_posted = job.find('span', class_ = "sim-posted").text.replace("\t\n","").strip().replace("Work from Home \n\t\t","")
  links_to_apply = job.header.h2.a['href']

  job_list =[[company_name,skills_req,date_posted,links_to_apply]]
  job_list_df = pd.DataFrame(job_list, columns =['Company', 'Skills', 'Date posted', 'Apply link'])

  empty_df = pd.concat([empty_df,job_list_df])
empty_df
```

```python
empty_df_10pages = pd.DataFrame( columns =['Company', 'Skills', 'Date posted', 'Apply link'])

for i in range(1,11):
  weblink = 'https://www.timesjobs.com/candidate/job-search.html?from=submit&luceneResultSize=25&txtKeywords=data%20science&cboWorkExp1=0&postWeek=60&searchType=personalizedSearch&actualTxtKeywords=Data%20Science&searchBy=0&rdoOperator=OR&txtLocation=Bengaluru/%20Bangalore&pDate=I&sequence='+str(i)+'&startPage=1'
  html_text = requests.get(weblink).text
  soup = BeautifulSoup(html_text, 'lxml')
  jobs = soup.find_all('li', class_ ="clearfix job-bx wht-shd-bx")

  for job in jobs:
    company_name = job.find('h3',class_="joblist-comp-name").text.replace("\t\n","").replace("(More Jobs)", "").strip()
    skills_req = job.find('span',class_="srp-skills").text.replace("\t\n","").strip()
    date_posted = job.find('span', class_ = "sim-posted").text.replace("\t\n","").strip().replace("Work from Home \n\t\t","")
    links_to_apply = job.header.h2.a['href']

    job_list =[[company_name,skills_req,date_posted,links_to_apply]]
    job_list_df = pd.DataFrame(job_list, columns =['Company', 'Skills', 'Date posted', 'Apply link'])

    empty_df_10pages = pd.concat([empty_df_10pages,job_list_df])

empty_df_10pages
```
